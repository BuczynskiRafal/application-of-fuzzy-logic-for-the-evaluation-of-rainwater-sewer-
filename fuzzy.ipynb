```python
# import plotly.io as pio
# pio.renderers.default = 'browser'
import numpy as np
import plotly.graph_objects as go

hydraulic_values = np.arange(0, 1.05, 0.05)
technical_values = np.arange(0, 1.05, 0.05)
hydraulic_values, technical_values = np.meshgrid(hydraulic_values, technical_values)
pred_val = np.zeros(shape=(len(technical_values), len(hydraulic_values[0])))

for i in range(len(hydraulic_values[0])):
    for j in range(len(technical_values)):
        sewer_simulation.input['hydraulic_performance'] = hydraulic_values[0, i]
        sewer_simulation.input['technical_condition'] = technical_values[j, 0]
        sewer_simulation.input['operational_condition'] = 0.4
        sewer_simulation.compute()
        pred_val[j][i] = sewer_simulation.output['consequence']

fig = go.Figure(data=[go.Surface(x=hydraulic_values, y=technical_values, z=pred_val)])
fig.update_layout(title='Consequence evaluation using the fuzzy logic model',
                  scene=dict(
                      xaxis_title='Hydraulic Performance',
                      yaxis_title='Technical Condition',
                      zaxis_title='Consequence evaluation'),
                  autosize=True,
                  width=900, height=900)

fig.show()
```
